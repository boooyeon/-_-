#**인공지능 1주차 과제**

##**<언어, 음성, 이미지, 자율주행> 분야의 서비스(제품)속 적용된 인공지능 기술의 분석**

###**1. 언어**
인공지능 분야에서 사람의 언어를 학습시키려는 노력은 계속되고 있다. 인공지능이 사람의 언어를 이해하기 위해서는 *자연어 처리(natural language processing)*과정을 거쳐야 한다. 자연어 처리란, 인간이 발화하는 언어 현상을 기계적으로 분석해 컴퓨터가 이해할 수 있는 형태로 만드는 것이다.
자연어 처리를 활용한 서비스로 쉽게는 *챗봇(chatbot)*을 예로 들 수 있다. 이미 많은 곳에서 챗봇 서비스를 도입하고 있다. 나 역시 챗봇 서비스로 간단한 정보(학식, 학사일정 등)들을 손쉽게 알아 내고 있어서 이미 생활 속에 깊숙히 자리 잡은 익숙한 기술이다.
언어 번역에 있어서는 신경망 번역을 통해 인공지능(AI)이 스스로 빅데이터를 학습해 문장 전체를 맥락에 맞게 번역한다. 기존 AI 번역은 단어를 하나하나 해석해 매끄럽지 못한 반면, 신경망 번역 기술은 방대한 데이터를 통해 전체 문맥을 파악해 실제 생활에서 쓰이는 의미를 최대한 살리는 장점을 가지고 있다.
인간의 말을 이해하고 해석하는 것을 넘어서 직접 언어를 구사하기도 한다. 최근에 엔씨소프트와 연합뉴스는 AI 미디어 공동연구의 결과물 중 하나로 AI 날씨 기사를 선보였다. 머신러닝 NLP 기술로 자동 작성한 기사를 내보였는데, 이 기술은 사람이 일반적으로 쓰는 말,  곧 자연어를 기계가 이해하고 구사하도록 하는 기술이라고 한다. AI 스피커 등에도 쓰인다고 한다. 여기서, 인공지능의 언어 분야가 많이 발달 했다는 점을 알 수 있는게 지금까지 국내 로봇 기사는 대부분 증시 시황, 스포츠 경기 전적 등 정형 데이터를 미리 정한 틀에 넣어 만드는 방식이었다. 따라서, 이 방식은 일기예보처럼 데이터가 복잡하고 문장 유형이 다양한 기사에는 적용하기 어렵기 때문이다. 
또한, IBM이 개발한 토론 전용 인공지능(AI) *'프로젝트 디베이터(Project Debater)'*를 인간과의 토론으로 인류의 사고력과 문제 해결력을 발전시키고자 개발하엿다. 신문과 학술자료에 실린 100억 개의 문장을 학습해 지식을 축적했고, 이 데이터를 기반으로 명확한 논리를 전달할 수 있으며 수준급의 문장 조합 실력을 자랑한다. 심지어 토론 중 농담을 하기도 하며 때로는 단호하고 거칠게 응수하기도 한다고 한다.
인공지능이 토론의 영역까지 들어왔다는 것은 그만큼 인간의 언어를 이해하고 해석해서 논리적으로 구현할 수 있는 수준까지 도달했다는 것을 의미해 언어 분야에 있어서 엄청난 발전을 했다는 것을 알 수 있었다.

##**2. 음성**
처음에 단순한 명령 정도만 알아듣는 음성 비서로 시작했던 기술은 이후에 더욱 다양한 분야에 적용되면서 빠르게 발전하고 있다. 인공지능 스피커가 등장 했고, 다양한 스마트 가전 기기에 음성인식 기능이 적용되고 있다. 스마트폰 대부분에 음성 인식 서비스가 탑재되어있다.(애플은 시리(siri), 삼성은 빅스비, 구글은 구글 어시스턴트) 음성 인식 서비스 역시 우리 주변에서 충분히 볼 수 있는 기술이다. 말 한마디로 날씨 예보는 물론 좋아하는 음악, 사고싶은 물건에 대한 쇼핑 정보를 알아낼 수 있을 정도로 음성인식 기술은 이미 우리 생활에 깊이 스며들었고, 널리 사용되고 있다.
이러한 음석인식 기술은 인공지능(AI)기술과 사물인터넷(IoT) 기술과 맞물려 다양한 분야로 점차 확대될 전망이다.
음성 인식 기술이 가장 활발하게 사용되고 있는 분야는 스피커라고 볼 수 있다. 국내에서는 SK텔레콤의 누구, KT의 기가지니, 카카오의 카카오 미니 등이 있다.
이들은 특정 단어로 스피커의 기능을 활성화한 후 다양한 실생활 속 필요한 정보를 얻을 수 있고, 다른 가전제품들과 연동하는 기술을 지니고 있다.
자동차 분야에서 역시 음성 인식 기술을 볼 수 있다. 
특히, 최근에는 AI가 결합된 음성인식 기술이 단순히 음성을 알아듣고 텍스트로 변환하는 단계를 넘어, 화자가 사용하는 단어와 목소리 톤 등을 분석해 감정까지 잡아낸다고 한다.
많은 발전을 이룬 음성인식 기술이지만, 아직도 한계는 존재한다. 사람마다 발음, 억양, 톤이 다르고, 또 일상 대화에서는 문법에 어긋난 문장도 많고 주변의 소음도 섞이기 때문이다. 이러한 한계를 극복하기 위해서 더 많은 데이터를 확보해야한다. 주변 소음이 담겨있는 음성이나 우리가 일상 생활에서 쓰는 구어체 음성, 의미는 통하지만 문법에 맞지 않는 문장에 대한 음성 등 다양한 음성 언어 데이터를 확보해야 인공지능의 음성인식 기술 성능을 향상시킬 수 있다. 즉, 방대한 양의 정확도 높은 데이터가 필요하다.

##**3. 이미지**
음성을 인식하는 것에서 한 단계 더 나아가 보는 영역에도 인공지능이 존재한다.
AI 이미지 분석 기술은 사물을 보고 스스로 인식한다. 렌즈를 갖다 대면 알아서 피사체를 분석해주는 AI카메라가 그 예이다. 이미지 분석 기술이 인공신경망 기술인 딥러닝을 통해 사람의 시각처럼 높은 정확도와 다양한 사물 인식이 가능해졌다.
실생활에서 찾아볼 수 있는 AI 이미지 분석 기술로는 인물이나 피사체의 움직임을 정교하게 인식하여 촬영하는 스마트폰의 카메라, 사진을 자동으로 분류해주는 구글 포토, 카메라를 비추면 어떤 사물인지 알려주는 구글 렌즈 등이 있다.
AI 이미지 분석은 다양한 산업 분야에 활용할 수 있다. 스마트 팩토리 같은 경우 사람이 일일이 확인하는 것보다 오류 발생 확률을 낮추고 불량 판정률을 높일 수 있다.
보안검색대에서는 사람이 하나하나 판별할 필요 없이 보다 정확하게 촬영 이미지를 분석하여 반.출입 불허 물품을 자동으로 검색할 수 있다. 또한, AI 무인 매장은 계산대 없이 고객이 물건을 가지고 나오면 저절로 결제가 된다. 이러한 무인 매장은 천장에 설치된 34대의 스마트 카메라와 진열대에 설치된 300여개의 무게 감지 센서 덕분에 사람이 아닌 AI가 매장을 관리할 수 있다. 물론 아직 이미지 인식에도 많은 한계가 존재한다. AI 이미지 분석의 핵심 기술은 '노이즈(noise)를 얼마나 판별해 내느냐 하는 것이다. 또한, 객체의 구도와 모양, 그리고 색이 유사하여 이미지 분석을 혼란스럽게 만든 경우 이미지를 잘못 인식하는 문제가 있다. 


##**4. 자율 주행**
자동차 분야에서도 인공지능은 많은 영향을 미치고 있다. 작고 강력한 컴퓨터와 GPS 시스템, AI 개발과 함께 자율주행 분야에서 많은 발전이 이루어지고 있다. 다양한 운전 조건에서 자동차의 제어 능력을 높이기 위해 인공지능(AI)이 사용된다.또한, 자동차 산업에서 자주 회자되는 단어가 연결성(Connectivity)이다. 사람들은 정보 및 다른 사람들과 연결하기를 원하고, 자동차 제조업체들은 이러한 작업을 수행할 수 있도록 AI를 활용한다. 
AI 기반의 인포테인먼트 시스템은 운전자와 승객이 음성 명령을 통해 이메일을 송수신하고 인터넷을 검색하며 스마트폰 앱을 수행할 수 있다. 일부 차량에는 곧 자체 Wi-Fi 핫스팟이 포함될 것이다.
이러한 AI가 제공하는 가장 중요한 헤택은 안정성 향상이다. 자동차가 서로 통신하면서 각 차량이 주변의 다른 차량과 정보를 교환해 사고를 줄일 수 있다. 연결성의 고도화 추세는 AI/자동차 분야에 진출하고자 하는 개발자들에게 새로운 기회를 제공한다.
자율주행차의 도입은 기존 교통체계와 자동차 산업 생태계를 새롭게 변화시킬 것이다. 이에 따라 당명한 문제가 있는데, 첫째는 소프트웨어 트레이닝 및 검증에 관한 과제다. 시스템이 트레이닝 후 학습한 내용을 프로그래머가 검증하기 어렵기 때문에 소프트웨어 성능을 평가하기 위한 몇가지 시뮬레이션이 필요하다. 두 번째는 보안 위험과 데이터 소유권에 대한 과제다. 해킹의 문제로부터 해결책을 가져야한다.마지막으로 자율주행과 관련한 법적 문제가 있다.

#**다양한 인공지능 분야에서 가장 흥미있는 분야와 앞으로의 목표**
모든 분야에서의 인공지능 활용이 매력이 있지만 나는 그중에서도 특히 음성, 이미지 분야와 관련해서 깊이 있게 알아보고 싶다. 
최근 MBC 다큐멘터리 '너를 만났다'에서 가상현실(VR)과 인공지능(AI) 기술로 세상을 떠난 아이를 되살리는 프로젝트를 진행하는 것을 봤다. 아이의 목소리를 구현하기 위해 '인공신경망 기반 기계학습(딥러닝)' 기술을 사용하였는데, 생전에 저장해 둔 1분 남짓의 음성 데이터에 5명 또래 아이의 목소리를 각각 800문장 이상씩을 더빙하고 이를 AI로 재구성 하였다고한다. 평소에 '잊혀짐'과 '그리움'에 있어서 고민을 많이 해왔다. 그에 따라 자연스럽게 기존의 보편적이고 획일화된 추모방식에 대해서 의문을 품은 나는 기존의 사진이나 동영상으로 죽은 사람을 그리워했듯이 이러한 기술로도 사람의 마음을 위로할 수 있을 거라고 생각이 들었다. 물론 부작용이나 윤리적인 관점에서의 문제도 고려를 해야겠지만, 인공지능이 단순히 힘들거나 귀찮은 일을 대신해줘 인간의 삶을 편리하고 윤택하게 해주는 것을 넘어서 인간의 복잡한 감정까지도 어루만질 수 있는 영역으로 발전하는 것을 공부하는 것이 앞으로 나의 학습 목표이다.
