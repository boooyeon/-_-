##**<언어, 음성, 이미지, 자율주행> 분야의 서비스(제품)속 적용된 인공지능 기술 및 분석**

###**1. 언어**
인공지능 분야에서 사람의 언어를 학습시키려는 노력은 계속되고 있다. 인공지능이 사람의 언어를 이해하기 위해서는 *자연어 처리(natural language processing)*과정을 거쳐야 한다. __자연어 처리란, 인간이 발화하는 언어 현상을 기계적으로 분석해 컴퓨터가 이해할 수 있는 형태로 만드는 것이다.__
자연어 처리를 활용한 서비스로 쉽게는 *챗봇(chatbot)*을 예로 들 수 있다. 이미 많은 곳에서 챗봇 서비스를 도입하고 있다. 나 역시 카카오 챗봇 서비스로 간단한 정보(학식, 학사일정 등)들을 손쉽게 알아 내고 있어서 이미 생활 속에 깊숙히 자리 잡은 익숙한 기술이다.
또한, 네이버는 업계에서 가장 먼저 선도적으로 NLP의 중요성을 인식하고 개발과 투자로 축적한 기술력과 서비스 노하우를 바탕으로 네이버 검색이 자연어 처리 분야에서 검색 이용자의 의도를 더욱 잘 이해하는 검색으로 진화하고 있다. 
또 모바일 상에서의 검색이 일상화된 이용자들을 위해 AI 기술 기반 검색어 교정 시스템인 ‘AIQSpell’ 개발에 힘써왔다. 딥러닝을 비롯한 최신의 AI 기술을 활용해 기존의 검색어 교정 시스템을 대체했으며, 최근 이에 대한 유의미한 성과가 나타나고 있다.
가장 최근의 글로벌 NLP 솔루션에서는 최첨단 AI 언어모델 중 하나인 구글 버트(BERT)를 한 시간 내에 훈련시키고, 2밀리초(ms) 만에 AI 인퍼런스를 완료한 최초의 모델로 실시간 대화형 인공지능(AI)을 통해 기업이 고객과 보다 자연스럽게 소통할 수 있는 엔비디아 '언어 이해 모델'을 꼽을 수 있다.


언어 번역에 있어서는 신경망 번역을 통해 인공지능(AI)이 스스로 빅데이터를 학습해 문장 전체를 맥락에 맞게 번역한다. 기존 AI 번역은 단어를 하나하나 해석해 매끄럽지 못한 반면, __신경망 번역 기술은 방대한 데이터를 통해 전체 문맥을 파악해 실제 생활에서 쓰이는 의미를 최대한 살리는 장점을 가지고 있다.__
카카오는 2018년 말부터 딥러닝 기반 형태소(形態素, morpheme) 분석기 '카이(khaiii)'를 오픈소스로 제공하고 있다. 딥러닝을 통해 학습한 데이터를 활용해 형태소를 분석하는 모델이다. 딥러닝 기술 중 하나인 콘볼루션 신경망(CNN, Convolutional Neural Network)을 이용해 음절기반으로 형태소를 분석하는 방법을 채택했다.
세종 코퍼스를 기반으로 데이터의 오류를 수정하고 카카오에서 자체 구축한 데이터를 추가해 85만 문장, 1003만 어절의 데이터를 학습하여 정확도를 높였다. 또 딥러닝 과정에서 C++ 언어를 적용해 일반적으로 딥러닝에 쓰이는 GPU(그래픽처리장치)를 사용하지 않고도 빠른 분석 속도를 구현했다.
LG CNS는 국내 처음이자 유일하게 AI의 언어 이해를 위한 AI 학습용 표준데이터 '코쿼드 2.0(KorQuAD 2.0)'를 공개하고 국내 AI 업계에 무료로 개방했다.

이렇게 인간의 말을 이해하고 해석하는 것을 넘어서 직접 언어를 구사하기도 한다. 최근에 엔씨소프트와 연합뉴스는 AI 미디어 공동연구의 결과물 중 하나로 AI 날씨 기사를 선보였다. 머신러닝 NLP 기술로 자동 작성한 기사를 내보였는데, 이 기술은 사람이 일반적으로 쓰는 말,  곧 자연어를 기계가 이해하고 구사하도록 하는 기술이라고 한다. AI 스피커 등에도 쓰인다고 한다. 여기서, 인공지능의 언어 분야가 많이 발달 했다는 점을 알 수 있는게 지금까지 국내 로봇 기사는 대부분 증시 시황, 스포츠 경기 전적 등 정형 데이터를 미리 정한 틀에 넣어 만드는 방식이었다. 따라서, 이 방식은 일기예보처럼 데이터가 복잡하고 문장 유형이 다양한 기사에는 적용하기 어렵기 때문이다. 

또한, IBM이 개발한 토론 전용 인공지능(AI) *'프로젝트 디베이터(Project Debater)'*를 인간과의 토론으로 인류의 사고력과 문제 해결력을 발전시키고자 개발하엿다. 신문과 학술자료에 실린 100억 개의 문장을 학습해 지식을 축적했고, 이 데이터를 기반으로 명확한 논리를 전달할 수 있으며 수준급의 문장 조합 실력을 자랑한다. 심지어 토론 중 농담을 하기도 하며 때로는 단호하고 거칠게 응수하기도 한다고 한다.
__인공지능이 토론의 영역까지 들어왔다는 것은 그만큼 인간의 언어를 이해하고 해석해서 논리적으로 구현할 수 있는 수준까지 도달했다는 것을 의미해 언어 분야에 있어서 엄청난 발전을 했다는 것을 알 수 있었다.__

##**2. 음성**
처음에 단순한 명령 정도만 알아듣는 음성 비서로 시작했던 기술은 이후에 더욱 다양한 분야에 적용되면서 빠르게 발전하고 있다. 인공지능 스피커가 등장 했고, 다양한 스마트 가전 기기에 음성인식 기능이 적용되고 있다. 스마트폰 대부분에 *음성 인식 서비스*가 탑재되어있다.(애플은 시리(siri), 삼성은 빅스비, 구글은 구글 어시스턴트) 음성 인식 서비스 역시 우리 주변에서 충분히 볼 수 있는 기술이다. __말 한마디로 날씨 예보는 물론 좋아하는 음악, 사고싶은 물건에 대한 쇼핑 정보를 알아낼 수 있을 정도로 음성인식 기술은 이미 우리 생활에 깊이 스며들었고, 널리 사용되고 있다.__

음성인식 가상비서 플랫폼 중 대표적으로 아마존의 '알렉사(Alexa)'와 구글의 'HeyGoogle'이 있다. 인공지능 가상비서는 4차산업혁명의 대표기술 중 하나로 소비자와 만나는 접점이 되고 있다. 기능으로 궁금한 것을 말로 질문하면 바로 알려주고 쇼핑, 예매까지 가능하다.
__인공지능 가상비서는 내 친구에서, 스마트홈 연계, B2B 모델로 진화하고 있고 스피커 형태에서 디스플레이형 제품으로 확장되고 있다.__

이러한 음석인식 기술은 인공지능(AI)기술과 사물인터넷(IoT) 기술과 맞물려 다양한 분야로 점차 확대될 전망이다.음성 인식 기술이 가장 활발하게 사용되고 있는 분야는 스피커라고 볼 수 있다. 주요 거대 기술기업은 모두 인공지능 스피커를 내놓고 있따. 스마트스피커 선도 회사인 아마존은 2014년말 에코(Echo) 스피커를 최초 출시한 이래, 저가형 제품인 Echo dot, 카메라로 본인에 적합한 의상을 골라주는 Echo Loock, 음성과 디스플레이가 결합된 서비스를 제공하는 Echo Show 등 새로운 디바이스를 잇따라 선보이고 있다. 구글의 경우 AI스타트업인 Api.ai사 인수를 기반으로 2016년 10월 구글홈(Google Home)스피커를 출시하였고, 애플은 2018년 시리가 탑재된 홈파드(HomePod)를 출시하였다. 중국도 중국어 음성인식 회사인 iFlyTec사가 딩동 스마트스피커를 2016년 말 출시하였고 화웨이에서도 제품을 선보였다. 국내에서도 주요 통신사 및 인터넷 기업을 중심으로 인공지능 스피커를 출시하였다. 2016년 9월  SK텔레콤의 누구, KT의 기가지니, 네이버는 자체 AI플랫폼인 클로버 기반의 웨이브, 캐릭터를 반영한 프렌즈를 카카오는 카카오톡과 연계된 카카오 미니등이 있다. 삼성전자 역시 음성 비서 빅스비를 탑재한 스피커를 2018년에 출시하였다. 이들은 특정 단어로 스피커의 기능을 활성화한 후 다양한 실생활 속 필요한 정보를 얻을 수 있고, 다른 가전제품들과 연동하는 기술을 지니고 있다.

자동차 분야에서 역시 음성 인식 기술을 볼 수 있다. 특히, 최근에는 AI가 결합된 음성인식 기술이 단순히 음성을 알아듣고 텍스트로 변환하는 단계를 넘어, 화자가 사용하는 단어와 목소리 톤 등을 분석해 감정까지 잡아낸다고 한다.

많은 발전을 이룬 음성인식 기술이지만, 아직도 한계는 존재한다. 사람마다 발음, 억양, 톤이 다르고, 또 일상 대화에서는 문법에 어긋난 문장도 많고 주변의 소음도 섞이기 때문이다. 이러한 한계를 극복하기 위해서 더 많은 데이터를 확보해야한다. 주변 소음이 담겨있는 음성이나 __우리가 일상 생활에서 쓰는 구어체 음성, 의미는 통하지만 문법에 맞지 않는 문장에 대한 음성 등 다양한 음성 언어 데이터를 확보해야 인공지능의 음성인식 기술 성능을 향상시킬 수 있다. 즉, 방대한 양의 정확도 높은 데이터가 필요하다.__

##**3. 이미지**
음성을 인식하는 것에서 한 단계 더 나아가 보는 영역에도 인공지능이 존재한다.
AI 이미지 분석 기술은 사물을 보고 스스로 인식한다. 렌즈를 갖다 대면 알아서 피사체를 분석해주는 *AI카메라*가 그 예이다. __이미지 분석 기술이 인공신경망 기술인 딥러닝을 통해 사람의 시각처럼 높은 정확도와 다양한 사물 인식이 가능해졌다__ 실생활에서 찾아볼 수 있는 AI 이미지 분석 기술로는 인물이나 피사체의 움직임을 정교하게 인식하여 촬영하는 스마트폰의 카메라, 사진을 자동으로 분류해주는 구글 포토, 카메라를 비추면 어떤 사물인지 알려주는 구글 렌즈 등이 있다.

AI 이미지 분석은 다양한 산업 분야에 활용할 수 있다. __스마트 팩토리 같은 경우 사람이 일일이 확인하는 것보다 오류 발생 확률을 낮추고 불량 판정률을 높일 수 있다.__ 보안검색대에서는 사람이 하나하나 판별할 필요 없이 보다 정확하게 촬영 이미지를 분석하여 반.출입 불허 물품을 자동으로 검색할 수 있다.

또한, *AI 무인 매장*은 계산대 없이 고객이 물건을 가지고 나오면 저절로 결제가 된다. 이러한 무인 매장은 천장에 설치된 34대의 스마트 카메라와 진열대에 설치된 300여개의 무게 감지 센서 덕분에 사람이 아닌 AI가 매장을 관리할 수 있다.
신세계가 한국형 '아마존고'를 선보인다. 스마트폰 애플리케이션(앱)을 통해 마트에서 물건을 고른 뒤 따로 계산하는 과정을 거치지 않고 그냥 걸어 나가면 자동으로 계산이 되는 '저스트 워크 아웃(Just Walk Out)' 기술을 국내 최초로 상용화했다. 저스트 워크 아웃 방식의 무인 매장이 가능하려면 각종 최첨단 기술이 뒷받침돼야 한다. 아마존고의 경우 고객이 매장에서 쇼핑하는 동안 자율주행기술이 적용된 센서가 부착된 원형 카메라가 천장에서 고객을 따라다니며 진열대에 놓인 상품을 집어드는 동작을 분석한다. 이렇게 분석한 결과를 바탕으로 고객이 쇼핑을 끝내고 매장에서 나갈 때 실제 쇼핑을 한 제품에 대한 결제만 자동으로 하는 방식이다. 구체적인 이용 방법으로는 마트 이용 고객이 SSG페이나 이마트24 모바일 앱을 통해 이마트24 무인매장 입장을 위한 QR코드를 생성한 뒤, 매장 문에 설치된 게이트에 접촉하면 입장할 수 있다. 이후 아마존고에서처럼 자유롭게 쇼핑한 뒤 매장을 나올 때 스마트폰에 설치된 SSG페이나 이마트24 모바일 앱을 통해 자동으로 결제가 이뤄진다. 셀프계산대도 필요없는 완전 무인매장이다.

물론 아직 이미지 인식에도 많은 한계가 존재한다. __AI 이미지 분석의 핵심 기술은 '노이즈(noise)를 얼마나 판별해 내느냐 하는 것이다.__ 또한, 객체의 구도와 모양, 그리고 색이 유사하여 이미지 분석을 혼란스럽게 만든 경우 이미지를 잘못 인식하는 문제가 있다. 

##**4. 자율 주행**
자동차 분야에서도 인공지능은 많은 영향을 미치고 있다. 작고 강력한 컴퓨터와 GPS 시스템, AI 개발과 함께 *자율주행분야*에서 많은 발전이 이루어지고 있다. 다양한 운전 조건에서 자동차의 제어 능력을 높이기 위해 인공지능(AI)이 사용된다.

또한, __자동차 산업에서 자주 회자되는 단어가 연결성(Connectivity)이다.__ 사람들은 정보 및 다른 사람들과 연결하기를 원하고, 자동차 제조업체들은 이러한 작업을 수행할 수 있도록 AI를 활용한다. AI 기반의 인포테인먼트 시스템은 운전자와 승객이 음성 명령을 통해 이메일을 송수신하고 인터넷을 검색하며 스마트폰 앱을 수행할 수 있다. 일부 차량에는 곧 자체 Wi-Fi 핫스팟이 포함될 것이다.
이러한 AI가 제공하는 가장 중요한 헤택은 안정성 향상이다. 자동차가 서로 통신하면서 각 차량이 주변의 다른 차량과 정보를 교환해 사고를 줄일 수 있다. 연결성의 고도화 추세는 AI/자동차 분야에 진출하고자 하는 개발자들에게 새로운 기회를 제공한다.

자동차가 스스로 차선과 장애물, 보행자 등을 인식해 운행하려면 주변 영상을 빠르게 관측해 해석할 수 있어야 한다. 하지만 아직 현실은 완전자율주행과 거리가 있다. 현재 자율주행차는 제한된 환경에서 짧은 시간 동안 사람의 간섭 없이 운행이 가능한 정도(3단계 자율주행)다. 
우리나라에서는 현대자동차와 기아자동차는 운전자의 주행성향에 맞는 부분 자율주행을 구현하는 기술인 SCC-ML을 세계 최초로 개발했다.
스마트 크루즈 컨트롤 SCC은 앞차와의 거리를 일정하게 유지해 운전자가 설정한 속도로 자율주행을 해주는 기능으로 ADAS(첨단 운전자보조 시스템)의 주요 기술 중 하나다. SCC-ML은 여기에 AI기술을 더해 운전자의 주행성향을 차가 스스로 학습해 SCC 작동 시 운전자와 거의 흡사한 패턴으로 자율주행을 해준다. 이번 현대차·기아차가 독자 개발한 SCC-ML의 원리는 먼저 전방카메라, 레이더 등의 센서가 다양한 운전상황에서 발생되는 정보를 지속적으로 수집해 ADAS의 두뇌격인 제어컴퓨터로 보낸다. 제어컴퓨터는 입력된 정보로부터 운전자의 주행습관을 추출해 종합적인 주행성향을 파악한다. 이때 인공지능 기술 중 하나인 머신러닝 알고리즘이 적용된다.
주행성향에 대한 정보는 센서를 통해 계속 업데이트되기 때문에 운전자의 최근 성향을 반영할 수 있다. 또 안전운전을 크게 벗어난 주행성향은 따르지 않도록 설정돼 있어서 신뢰성을 높였다. 또 SCC-ML은 자동 차로 변경 기능을 포함하고 있는 HDA II와 함께 적용돼 자율주행 레벨 2을 넘어선 레벨 2.5 수준을 구현한다.

__자율주행차의 도입은 기존 교통체계와 자동차 산업 생태계를 새롭게 변화시킬 것이다.__ 이에 따라 당명한 문제가 있는데, 첫 번째는 소프트웨어 트레이닝 및 검증에 관한 과제다. 시스템이 트레이닝 후 학습한 내용을 프로그래머가 검증하기 어렵기 때문에 소프트웨어 성능을 평가하기 위한 몇가지 시뮬레이션이 필요하다. 두 번째는 보안 위험과 데이터 소유권에 대한 과제다. 해킹의 문제로부터 해결책을 가져야한다.마지막으로 자율주행과 관련한 법적 문제가 있다.

#**다양한 인공지능 분야에서 가장 흥미있는 분야와 앞으로의 목표**
모든 분야에서의 인공지능 활용이 매력이 있지만 나는 그중에서도 특히 음성, 이미지 분야와 관련해서 깊이 있게 알아보고 싶다. 
최근 MBC 다큐멘터리 '너를 만났다'에서 가상현실(VR)과 인공지능(AI) 기술로 세상을 떠난 아이를 되살리는 프로젝트를 진행하는 것을 봤다. 아이의 목소리를 구현하기 위해 '인공신경망 기반 기계학습(딥러닝)' 기술을 사용하였는데, 생전에 저장해 둔 1분 남짓의 음성 데이터에 5명 또래 아이의 목소리를 각각 800문장 이상씩을 더빙하고 이를 AI로 재구성 하였다고한다. 평소에 '잊혀짐'과 '그리움'에 있어서 고민을 많이 해왔다. 그에 따라 자연스럽게 기존의 보편적이고 획일화된 추모방식에 대해서 의문을 품은 나는 사진이나 동영상으로 죽은 사람을 그리워하고 추모하는 것을 넘어서 인공지능 기술을 접목시켜 색다른 방법으로도 사람의 마음을 위로할 수 있을 거라고 생각이 들었다. 물론 부작용이나 윤리적인 관점에서의 문제도 고려를 해야겠지만, __인공지능이 단순히 힘들거나 귀찮은 일을 대신해줘 인간의 삶을 편리하고 윤택하게 해주는 것을 넘어서 인간의 복잡한 감정까지도 어루만질 수 있는 영역으로 발전하는 것을 공부하는 것이 앞으로 나의 학습 목표이다.__
